# What is the Zip Function?

The zip() function is used to combine multiple lists into pairs.
It is useful when working with related data that should be processed together.


```python
names = ["Tommy", "Fred", "Gail"]
scores = [85, 90, 88]

list(zip(names, scores))
```

# Why can it be Confusing?

At first, zip() can be confusing because it returns paired values instead of a single list.
Converting the result to a list makes the output easier to understand.


```python
for name, score in zip(names, scores):
    print(name, score)
```

# Summary

The zip() function helps combine related lists into pairs, making it easier to work with connected data.
This is especially helpful when looping through multiple lists at the same time.
